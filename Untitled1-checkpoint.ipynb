```python
import re

class PorterStemmer:

    def __init__(self):
        self.vowel = re.compile(r'[aeiou]')
        self.double_consonant = re.compile(r'([^aeiou])\1$')
        self.cvc_pattern = re.compile(r'[^aeiou][aeiou][^aeiouwxy]$')

    # ---------- Measure Function ----------
    def measure(self, stem):
        form = re.sub(r'[^aeiou]+', 'C', stem)
        form = re.sub(r'[aeiou]+', 'V', form)
        return form.count('VC')

    def contains_vowel(self, stem):
        return bool(self.vowel.search(stem))

    # ---------- Stage 1a ----------
    def stage1a(self, word):
        if word.endswith("sses"):
            return word[:-2]
        elif word.endswith("ies"):
            return word[:-2]
        elif word.endswith("ss"):
            return word
        elif word.endswith("s"):
            return word[:-1]
        return word

    # ---------- Stage 1b ----------
    def stage1b(self, word):
        if word.endswith("eed"):
            stem = word[:-3]
            if self.measure(stem) > 0:
                return stem + "ee"

        elif word.endswith("ed"):
            stem = word[:-2]
            if self.contains_vowel(stem):
                word = stem

        elif word.endswith("ing"):
            stem = word[:-3]
            if self.contains_vowel(stem):
                word = stem

        if word.endswith(("at","bl","iz")):
            word += "e"
        elif self.double_consonant.search(word) and not word.endswith(("l","s","z")):
            word = word[:-1]
        elif self.measure(word) == 1 and self.cvc_pattern.search(word):
            word += "e"

        return word

    # ---------- Stage 1c ----------
    def stage1c(self, word):
        if word.endswith("y"):
            stem = word[:-1]
            if self.contains_vowel(stem):
                return stem + "i"
        return word

    # ---------- Stage 2 ----------
    def stage2(self, word):
        rules = {
            "ational": "ate","tional": "tion","enci": "ence","anci": "ance",
            "izer": "ize","abli": "able","alli": "al","entli": "ent",
            "eli": "e","ousli": "ous","ization": "ize","ation": "ate",
            "ator": "ate","alism": "al","iveness": "ive","fulness": "ful",
            "ousness": "ous","aliti": "al","iviti": "ive","biliti": "ble"
        }

        for suf, rep in rules.items():
            if word.endswith(suf):
                stem = word[:-len(suf)]
                if self.measure(stem) > 0:
                    return stem + rep
        return word

    # ---------- Stage 3 ----------
    def stage3(self, word):
        rules = {
            "icate": "ic","ative": "","alize": "al",
            "iciti": "ic","ical": "ic","ful": "","ness": ""
        }

        for suf, rep in rules.items():
            if word.endswith(suf):
                stem = word[:-len(suf)]
                if self.measure(stem) > 0:
                    return stem + rep
        return word

    # ---------- Stage 4 ----------
    def stage4(self, word):
        suffixes = [
            "al","ance","ence","er","ic","able","ible","ant",
            "ement","ment","ent","ion","ou","ism","ate",
            "iti","ous","ive","ize"
        ]

        for suf in suffixes:
            if word.endswith(suf):
                stem = word[:-len(suf)]
                if self.measure(stem) > 1:
                    if suf == "ion":
                        if stem.endswith(("s","t")):
                            return stem
                    else:
                        return stem
        return word

    # ---------- Stage 5 ----------
    def stage5(self, word):
        if word.endswith("e"):
            stem = word[:-1]
            m = self.measure(stem)
            if m > 1 or (m == 1 and not self.cvc_pattern.search(stem)):
                word = stem

        if self.measure(word) > 1 and word.endswith("ll"):
            word = word[:-1]

        return word

    # ---------- Full Stem ----------
    def stem(self, word):
        word = word.lower()

        word = self.stage1a(word)
        word = self.stage1b(word)
        word = self.stage1c(word)
        word = self.stage2(word)
        word = self.stage3(word)
        word = self.stage4(word)
        word = self.stage5(word)

        return word


# ---------------- MAIN EXPERIMENT ----------------

stemmer = PorterStemmer()

input_file = "words.txt"
gold_file = "gold_roots.txt"
output_file = "stems.txt"

# Read words
with open(input_file) as f:
    words = [w.strip().lower() for w in f]

# Generate stems
predicted = [stemmer.stem(w) for w in words]

# Save stems
with open(output_file, "w") as f:
    for s in predicted:
        f.write(s + "\n")

# Read gold stems
with open(gold_file) as f:
    gold = [g.strip().lower() for g in f]

# Accuracy
total = min(len(predicted), len(gold))
correct = sum(1 for i in range(total) if predicted[i] == gold[i])

print("Accuracy =", round((correct/total)*100,2), "%")

# User sentence
text = input("Enter sentence: ")
tokens = text.split()
print("Stemmed:", " ".join(stemmer.stem(t) for t in tokens))
```
